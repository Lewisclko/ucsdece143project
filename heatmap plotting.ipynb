```python
#Chen-Lin Ko
#plotting 
'''
this part is to build up visualization on our data. the goal is to utilize cleaned data and plot into 
heat maps and bar charts for analysis. Modules are needed to achieve our goal. Modules incldues pandas,
gmaps, geopy, and matplotlib.

'''
import pandas as pd
import os
from geopy import geocoders

os.chdir('C:\Users\Lewis\Desktop\scorecard_data')
gn = geocoders.GeoNames(username='lewisclko')

fname = 'MERGED2014_15_PP.csv'
data=pd.read_csv(fname, usecols=['CITY','UGDS_WHITE'])

df = pd.DataFrame(data)
la_list = []
lo_list = []
race_list = []
#total would be 7703 rows
for i in range(50):
    name = df.loc[i,'CITY'] 
    name = name+',us'                         #ensure US citys
    la_list.append(gn.geocode(name)[1][0])    #latitude
    lo_list.append(gn.geocode(name)[1][1])    #longitude
    race_list.append(df.loc[i,'UGDS_WHITE'])  #race ratio data
    df1 = pd.DataFrame({'La':la_list, 'Lo':lo_list,'Ra':race_list})
df1
```

```python
locations =df1[['La','Lo']]
weights = df1['Ra']
import gmaps
gmaps.configure(api_key='AIzaSyC_BDsDDMiMe2ztcitSmC7XIOrk5tuFcS8') # Fill in with your API key
fig = gmaps.figure()
fig.add_layer(gmaps.heatmap_layer(locations, weights=weights,max_intensity=1, point_radius=10))
fig
```
